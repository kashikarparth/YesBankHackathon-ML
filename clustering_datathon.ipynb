Importing all the requirements for clustering addresses

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from IPython.display import display
```

Importing the dataset, and cleaning the obvious invalid entries.

```python
df = pd.read_csv("datathon.csv", sep = "|", delimiter = "|", keep_default_na = False, skipinitialspace = True)
df = df[df.current_address3 != "add3"]
df = df.drop(columns=['cust_id_token'])
df = df[df.current_city != "DUMMY"]
```

# The following section is for city based clustering to a database 

Writing a loop to generate the list of pin codes for a given city, to be further used to generate clustered

```python
city_name = "JAIPUR"
df_city = df[df.current_city == city_name]
pin_codes = np.unique(df_city["current_pin_code"])
pin_codes_masker = [x.isdigit() for x in pin_codes]
pin_codes = [i for (i, v) in zip(pin_codes, pin_codes_masker) if v]
```

Display metrics, to gauge computation times

```python
display(len(df_city["current_address3"]))
```

Writing a loop to loop over all the pin codes, cluster, and append to database for final generation

```python
df2 = pd.DataFrame()
counter = 0
for pin in pin_codes:
    counter = counter + 1
    df1 = df_city[df_city.current_pin_code == pin]
    address_list = list(df1["current_address3"])
    
    if not(np.count_nonzero(address_list)) or sum(len(i) for i in address_list) <= len(address_list) :
        df_temp = df1
        classes = []
        for i in range(len(address_list)):
            classes.append(0)
        df_temp['cluster number'] = classes
        df2 = df2.append(df_temp)
    
    if sum(len(i) for i in address_list) > len(address_list):
        vectorizer = TfidfVectorizer()
        X = vectorizer.fit_transform(address_list)
        Y = X.toarray()
        c = int(len(np.unique(Y,axis = 0)))
        print ("n_samples = " + str(c), "for pin code: "+ pin + "( " + str(counter)+"/" + str(len(pin_codes)) +" )")
        
        if(c == 1):
            df_temp = df1
            classes = 0
            df_temp['cluster number'] = classes
            df2 = df2.append(df_temp)
            
        else:
            n_clusters = int(np.maximum(2,int(c/1.5)))
            kmeans = KMeans(n_clusters=n_clusters).fit(Y)
            for i in range(n_clusters):
                classes = []
                tf = [x==i for x in kmeans.labels_]
                df_temp = df1[tf]
                for j in range(np.count_nonzero(tf)):
                    classes.append(i)
                df_temp['cluster number'] = classes
                df2 = df2.append(df_temp)
display(df2)
```

Outputting the final dataframe, saving it

```python
display(df2)
df2.to_csv(str(city_name) +"_df.csv")
```

Longest entry per cluster

```python
df3 = pd.DataFrame()
for pin in pin_codes:
    df1 = df2[df2.current_pin_code == pin]
    clusters_unique = list(set(list(df1["cluster number"])))
    clusters = list(df1["cluster number"])
    for cluster in clusters_unique:
        tf = [x==cluster for x in clusters]
        df_temp = df1[tf]
        adds= list(df_temp["current_address3"])
        adds = [[i.split()] for i in adds]
        ind = adds.index(max(adds))
        df3 = df3.append(df_temp.iloc[[ind]])
```

Outputting the final dataframe, saving it

```python
#df3 = df3.drop(columns =['cluster number'])
display(df3)
df3.to_csv(str(city_name) +"_df_CLUSTERHEADS.csv")
```

Taking a trial pin code, and tokenizing the addresses

Trying to use silhouette method to figure out best cluster number upto c/1.5

```python
score = 0
best_cluster_number = 0
for n_clusters in range(int(c/1.7),int(c/1.5)):
    kmeans = KMeans(n_clusters=n_clusters).fit(Y)
    tmp = silhouette_score (Y, kmeans.labels_, metric='euclidean')
    if score<tmp:
        score = tmp
        best_cluster_number = n_clusters
        print("Calculating for cluster number: "+ str(n_clusters) + "/" +str(int(c/1.5)))
```
